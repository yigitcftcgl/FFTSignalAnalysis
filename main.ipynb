```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 30  
Ts = 1 / fs
t = np.arange(0, 1, Ts)

f1 = 2
f2 = 5
```

```python
x1 = np.cos(2 * np.pi * f1 * t)
x2 = 3 * np.cos(2 * np.pi * f2 * t)

x3 = x1 + x2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label="x1(t) (f1 = 2 Hz)")
plt.plot(t, x2, label="x2(t) (f2 = 5 Hz)")
plt.plot(t, x3, label="x3(t) = x1(t) + x2(t)", linestyle="--")
plt.title("x1, x2, x3 zaman ekseninde sinyalleri")
plt.xlabel("Zaman (t)")
plt.ylabel("Genlik")
plt.legend()
plt.grid()
plt.show()
```
